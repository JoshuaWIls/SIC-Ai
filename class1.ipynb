```python
import cv2

input_path = r"C:\Users\student\Downloads\istockphoto-1446695356-640_adpp_is.mp4"
output_path = r"C:\Users\student\Downloads\reversed_video.mp4"

# Open the video file
cap = cv2.VideoCapture(input_path)

# Check if video opened successfully
if not cap.isOpened():
    print("Error opening video file")
    exit()

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # For mp4 output

frames = []

# Read all frames and store them
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)

cap.release()

print(f"Total frames read: {len(frames)}")

# Create a VideoWriter to save the reversed video
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Write frames in reverse order
for frame in reversed(frames):
    out.write(frame)

out.release()

print("Reversed video saved to:", output_path)
```

```python
import pandas as pd

# Given data
height = [30, 32, 34, 28, 38, 26, 40, 44, 36, 28, 40, 39, 37]
weight = [2.8, 3, 2.7, 3.6, 2.2, 3.8, 4.1, 3.1, 2.8, 2.9, 3.6, 3.9, 2.8]

df1 = pd.DataFrame({'height': height, 'weight': weight})

df1['y_pred'] = 0.1 * df1['height'] + 0.15

df1['loss']=(df1['y_pred']-weight)**2

print(df1)
print('\n')
print('MSE is',df1['loss'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

height = [30, 32, 34, 28, 38, 26, 40, 44, 36, 28, 40, 39, 37]
weight = [2.8, 3, 2.7, 3.6, 2.2, 3.8, 4.1, 3.1, 2.8, 2.9, 3.6, 3.9, 2.8]

df = pd.DataFrame({'height': height, 'weight': weight})

df['y_pred1'] = 0.1 * df['height'] + 0.15

df['y_pred2'] = 0.095 * df['height'] - 0.15

plt.figure(figsize=(8,5))
plt.plot(df['height'], df['weight'], 'o-', label='Actual weight', color='black')
plt.plot(df['height'], df['y_pred1'], 's--', label='Prediction 1: 0.1*height + 0.15', color='blue')
plt.plot(df['height'], df['y_pred2'], 'd--', label='Prediction 2: 0.095*height - 0.15', color='red')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Actual vs Predicted Weight')
plt.legend()
plt.grid(True)
plt.show()
```
