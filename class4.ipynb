```python
import math
word_vectors = {
    'alice': [0.1, 0.3, 0.5, 0.7],
    'wonderland': [0.2, 0.4, 0.6, 0.8],
    'machines': [0.9, 0.1, 0.3, 0.5],
    'rabbit': [0.3, 0.5, 0.7, 0.9]
}
def cosine_similarity(vec1, vec2):
    dot_product = sum(a*b for a, b in zip(vec1, vec2))
    magnitude1 = math.sqrt(sum(a*a for a in vec1))
    magnitude2 = math.sqrt(sum(b*b for b in vec2))
    if magnitude1 == 0 or magnitude2 == 0:
        return 0
    return dot_product / (magnitude1 * magnitude2)
def main():
    word1 = input("Enter first word: ").lower()
    word2 = input("Enter second word: ").lower()
    if word1 not in word_vectors or word2 not in word_vectors:
        print("Sorry, one or both words are not in the vocabulary.")
        return
    vec1 = word_vectors[word1]
    vec2 = word_vectors[word2]
    similarity = cosine_similarity(vec1, vec2)
    print(f"Cosine similarity between '{word1}' and '{word2}': {similarity:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
a = np.array([5, 2, 3])
b = np.array([-2, -1, 5])
c = np.array([1, 3, -2])

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

print(f"Cosine similarity between a and b: {cosine_similarity(a, b):.4f}")
x_perp, y_perp = 1, 0
z_perp = -(5*x_perp + 2*y_perp) / 3
perp_vector = np.array([x_perp, y_perp, z_perp])
print(f"Vector perpendicular to a: {perp_vector}")

k = 2
parallel_vector = k * a
print(f"Vector parallel to a (k=2): {parallel_vector}")

a_plus_2b = a + 2*b
print(f"a + 2b = {a_plus_2b}")

dot_bc = np.dot(b, c)
print(f"Dot product of b and c: {dot_bc}")
if dot_bc == 0:
    print(f"Any vector parallel to b is perpendicular to c, e.g., {b}")
else:
    print("No non-zero vector parallel to b is perpendicular to c.")

# Visualization

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

origin = np.array([0, 0, 0])

def draw_vector(v, color, label):
    ax.quiver(*origin, *v, color=color, label=label, arrow_length_ratio=0.1)

draw_vector(a, 'blue', 'a = (5, 2, 3)')
draw_vector(b, 'green', 'b = (-2, -1, 5)')
draw_vector(c, 'red', 'c = (1, 3, -2)')
draw_vector(perp_vector, 'purple', 'Perp to a')
draw_vector(parallel_vector, 'cyan', 'Parallel to a (2*a)')
draw_vector(a_plus_2b, 'orange', 'a + 2b')

all_vectors = np.array([a, b, c, perp_vector, parallel_vector, a_plus_2b])
max_val = np.max(np.abs(all_vectors))
ax.set_xlim([-max_val, max_val])
ax.set_ylim([-max_val, max_val])
ax.set_zlim([-max_val, max_val])

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Vector Visualizations')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
b = np.array([-2, -1, 5])
c = np.array([1, 3, -2])
dot_bc = np.dot(b, c)
print(f"Dot product of b and c: {dot_bc}")

if dot_bc == 0:
    print("Any scalar multiple of b is perpendicular to c.")
    v = b  
else:
    print("No non-zero vector parallel to b is perpendicular to c.")
    v = np.array([0, 0, 0])  
origin = np.array([0, 0, 0])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

def draw_vector(vec, color, label):
    ax.quiver(*origin, *vec, color=color, label=label, arrow_length_ratio=0.1)

draw_vector(b, 'blue', 'Vector b')
draw_vector(c, 'red', 'Vector c')
draw_vector(v, 'green', 'Vector parallel to b & perpendicular to c (if any)')

max_val = max(np.max(np.abs(b)), np.max(np.abs(c)), 1)
ax.set_xlim([-max_val, max_val])
ax.set_ylim([-max_val, max_val])
ax.set_zlim([-max_val, max_val])

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Vectors b, c, and vector parallel to b & perpendicular to c')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
king = np.array([0.8, 0.6, 0.9])
queen = np.array([0.7, 0.7, 0.85])
man = np.array([0.6, 0.5, 0.8])
woman = man - king + queen

print("king vector:", king)
print("queen vector:", queen)
print("man vector:", man)
print("woman vector (calculated):", woman)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origin = np.zeros(3)
def draw_vector(vec, color, label):
    ax.quiver(*origin, *vec, color=color, label=label, arrow_length_ratio=0.1)

draw_vector(king, 'blue', 'King')
draw_vector(queen, 'purple', 'Queen')
draw_vector(man, 'green', 'Man')
draw_vector(woman, 'red', 'Woman')

max_val = max(np.max(np.abs([king, queen, man, woman])), 1)
ax.set_xlim([-max_val, max_val])
ax.set_ylim([-max_val, max_val])
ax.set_zlim([-max_val, max_val])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Vector Analogy: king - queen = man - woman")
ax.legend()

plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np
mango_vector = model['mango']
joy_vector = model['joy']
anger_vector = model['anger']
happiness_vector = model['happiness']
vectors = np.array([mango_vector, joy_vector, anger_vector, happiness_vector])
print("Original shape:", vectors.shape) 
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)
print("Reduced vectors (2D):")
print(vectors_2d)
```

```python
import matplotlib.pyplot as plt

words = ['mango', 'joy', 'anger', 'happiness']

plt.figure(figsize=(8,6))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], color='blue')
for i, word in enumerate(words):
    plt.text(vectors_2d[i, 0] + 0.01, vectors_2d[i, 1] + 0.01, word, fontsize=12)

plt.title('2D PCA Projection of Word Vectors')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import numpy as np

words = ['mango', 'joy', 'anger', 'happiness']
pca = PCA(n_components=3)
vectors_3d = pca.fit_transform(vectors)

print("Reduced vectors (3D):")
print(vectors_3d)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(vectors_3d[:, 0], vectors_3d[:, 1], vectors_3d[:, 2], color='green')

# Add labels near points
for i, word in enumerate(words):
    ax.text(vectors_3d[i, 0], vectors_3d[i, 1], vectors_3d[i, 2], word, fontsize=12)

ax.set_title('3D PCA Projection of Word Vectors')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def f(x, y):
    return 3*x**2 - 4*x*y + 2*y**2 + 3*x - 6*y + 20
x = np.linspace(0, 3, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
x_min, y_min = 3/2, 3
z_min = f(x_min, y_min)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(x_min, y_min, z_min, color='r', s=100, label='Minimum point')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Surface plot of $f(x,y) = 3x^2 - 4xy + 2y^2 + 3x - 6y + 20$')
ax.legend()
plt.show()
```
